```python
class SmartLight:
    def __init__(self, light_name):
        self.light_name = light_name
        self.is_on = False   # Default OFF

    def turn_on(self):
        self.is_on = True
        print(self.light_name + " is ON")

    def turn_off(self):
        self.is_on = False
        print(self.light_name + " is OFF")


# Sample Input
light = SmartLight("Bedroom Light")
light.turn_on()   # Action: ON
```
